## Hint: 服飾

來源: DataPi 作者:Adrian Rosebrock

(https://codertw.com/%E7%A8%8B%E5%BC%8F%E8%AA%9E%E8%A8%80/516987/)

目標: 構建一個時尚影象的搜尋引擎。使用他的app，使用者可以上傳一張他們喜歡的服飾的圖片（如襯衫、裙子、褲子、鞋類），系統將會返回相似的物品並附上購買連結。()

評估指標: 服飾類別(襯衫、裙子、褲子、鞋類等)、顏色(紅、藍、黑等)、質地(棉、羊毛、絲、麻等)
